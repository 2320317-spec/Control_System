```python
import control as ct
import matplotlib.pyplot as plt
import numpy as np

# Define the Transfer Function H(s) = X2(s) / F(s)
# Based on the derived equations:
# Numerator: (s + 1)
# Denominator: (s^2 + 2s + 2)^2 - (s + 1)^2
# Expanding the denominator: (s^4 + 4s^3 + 8s^2 + 8s + 4) - (s^2 + 2s + 1)
# Resulting Denominator: s^4 + 4s^3 + 7s^2 + 6s + 3

num = [1, 1]
den = [1, 4, 7, 6, 3]

sys = ct.TransferFunction(num, den)

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# 1. Pole-Zero Map
ct.pole_zero_plot(sys, ax=ax1)
ax1.set_title('Pole-Zero Map')
ax1.grid(True, which='both', linestyle='--', alpha=0.5)

# 2. Step Response (The Graph)
t, y = ct.step_response(sys)
ax2.plot(t, y, lw=2, color='blue')
ax2.set_title('Step Response: $X_2(t)$')
ax2.set_xlabel('Time (seconds)')
ax2.set_ylabel('Amplitude')
ax2.grid(True, which='both', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()

# Display the poles and zeros in the console
print(f"System Poles: {ct.poles(sys)}")
print(f"System Zeros: {ct.zeros(sys)}")
```
