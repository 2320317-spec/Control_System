```python
# Install these if you haven't yet: !pip install openai langchain-community
from openai import OpenAI
from langchain_community.chat_message_histories import ChatMessageHistory

print("Libraries loaded successfully.")
```

```python
# Updated Cell 2
client = OpenAI(
    base_url="http://127.0.0.1:1234/v1",
    api_key="lm-studio"
)

MODEL_NAME = "deepseek-r1-0528-qwen3-8b" # Keep your loaded model

# This is the "Secret Sauce" to make it less robotic
SYSTEM_INSTRUCTION = {
    "role": "system", 
    "content": "You are a helpful assistant. Provide direct answers. Do NOT show your internal reasoning process or 'thinking' steps. Be brief and conversational."
}
```

```python
# Initialize/Reset the history object
history = ChatMessageHistory()
print("Conversation history has been initialized/cleared.")
```

```python
print("--- Chat Started (Type 'exit' to stop) ---")

while True:
    query = input("Type your message here: ").strip()

    if query.lower() == "exit":
        break
    if not query:
        continue

    print(f"\nUSER: {query}") 
    history.add_user_message(query)

    # Inject the System Instruction at the very top every time
    formatted_messages = [SYSTEM_INSTRUCTION] 
    
    for msg in history.messages:
        role = "user" if msg.type == "human" else "assistant"
        formatted_messages.append({"role": role, "content": msg.content})

    print("-" * 30)
    print("AI: ", end="", flush=True)

    try:
        stream = client.chat.completions.create(
            model=MODEL_NAME,
            messages=formatted_messages,
            stream=True,
            temperature=0.7 # A bit of "randomness" makes it sound more human
        )

        full_response = ""
        for chunk in stream:
            content = getattr(chunk.choices[0].delta, "content", None)
            if content:
                print(content, end="", flush=True)
                full_response += content
        
        history.add_ai_message(full_response)
        print("\n" + "-" * 30)

    except Exception as e:
        print(f"\n[Error]: {e}")
        break
```
