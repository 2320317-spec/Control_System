```python
import serial
import time

# Initialize connection on COM4
try:
    arduino = serial.Serial(port='COM4', baudrate=9600, timeout=1)
    time.sleep(2) # Buffer time for Arduino to reboot
    print("✅ Connected to Arduino on COM4")
except Exception as e:
    print(f"❌ Connection Error: {e}")
    # We use sys.exit or a flag here in a real app, 
    # but for this script, we'll just stop.

while True:
    command = input("Enter an angle (0-180) or 'q' to quit: ")

    if command.lower() == 'q':
        print("Closing connection...")
        break
    
    # --- Error Catching Starts Here ---
    try:
        # 1. Convert input to a whole number
        angle = int(command)
        
        # 2. Check if it's within the Servo's physical limits
        if 0 <= angle <= 180:
            # We send it as a string followed by a newline (\n)
            # The Arduino uses that newline to know the number ended
            arduino.write(f"{angle}\n".encode())
            
            # 3. Read back the Arduino's confirmation
            time.sleep(0.1)
            feedback = arduino.readline().decode('utf-8').strip()
            print(f"Arduino Status: {feedback}")
        else:
            print("⚠️ Error: Angle must be between 0 and 180.")
            
    except ValueError:
        # This catches letters, symbols, or empty spaces
        print("⚠️ Error: Invalid input. Please enter a whole number.")

arduino.close()
```
